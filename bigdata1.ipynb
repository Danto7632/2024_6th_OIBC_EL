```python
import requests

# 날짜 설정 및 API 키
date = '2024-10-24'
API_KEY = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJzdWIiOiJINXI3TXZGZm9IUGZnaEd0UGRBMk1pIiwiaWF0IjoxNzMwODA2NzcyLCJleHAiOjE3MzE1OTY0MDAsInR5cGUiOiJhcGlfa2V5In0.lWotiXYOoP-tEQ5W5B43dpLiMoRSj8l8zUgxrCjOLos'

# actual_weather API 호출
actual_weather = requests.get(
    f'https://research-api.solarkim.com/data/cmpt-2024/actual-weather/{date}',
    headers={'Authorization': f'Bearer {API_KEY}'}
).json()

# weather_forecast API 호출
weather_forecast = requests.get(
    f'https://research-api.solarkim.com/data/cmpt-2024/weather-forecast/{date}',
    headers={'Authorization': f'Bearer {API_KEY}'}
).json()

# actual_weather 데이터에서 필요한 목록 가져오기
actual_weather_1 = actual_weather.get('actual_weather_1', [])
actual_weather_2 = actual_weather.get('actual_weather_2', [])

# weather_forecast 데이터에서 weather_forecast_2 가져오기
weather_forecast_2 = weather_forecast.get('weather_forecast_2', [])

# ts를 기준으로 오름차순 정렬
sorted_weather_1 = sorted(actual_weather_1, key=lambda x: x['ts'])
sorted_weather_2 = sorted(actual_weather_2, key=lambda x: x['ts'])
sorted_forecast_2 = sorted(weather_forecast_2, key=lambda x: x['ts'])

# 최대 키 값을 기준으로 24로 나눈 인덱스 간격 계산
step_1 = len(sorted_weather_1) // 24
step_2 = len(sorted_weather_2) // 24
step_forecast = len(sorted_forecast_2) // 24

# 결과 출력
print("count | pressure | ground_press | for_ground_press")
print("-" * 50)

for count in range(1, 25):  # 1부터 24까지
    index_1 = count * step_1 - 1 if step_1 > 0 else 0
    index_2 = count * step_2 - 1 if step_2 > 0 else 0
    index_forecast = count * step_forecast - 1 if step_forecast > 0 else 0

    # actual_weather에서 pressure와 ground_press 추출
    pressure_value = sorted_weather_1[index_1]['pressure'] if index_1 < len(sorted_weather_1) else None
    ground_press_value = sorted_weather_2[index_2]['ground_press'] if index_2 < len(sorted_weather_2) else None

    # weather_forecast에서 for_ground_press 추출
    for_ground_press_value = sorted_forecast_2[index_forecast]['ground_press'] if index_forecast < len(sorted_forecast_2) else None

    # 출력
    print(f"{count:5d} | {pressure_value:8} | {ground_press_value:12} | {for_ground_press_value:16}")
```

```python
import requests

# 날짜 및 API 키 설정
date = '2024-11-04'
API_KEY = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJzdWIiOiJINXI3TXZGZm9IUGZnaEd0UGRBMk1pIiwiaWF0IjoxNzMwODA2NzcyLCJleHAiOjE3MzE1OTY0MDAsInR5cGUiOiJhcGlfa2V5In0.lWotiXYOoP-tEQ5W5B43dpLiMoRSj8l8zUgxrCjOLos'

# actual_weather 데이터 요청
actual_weather = requests.get(
    f'https://research-api.solarkim.com/data/cmpt-2024/actual-weather/{date}',
    headers={'Authorization': f'Bearer {API_KEY}'}
).json()

# actual_weather 데이터에서 필요한 목록 추출
actual_weather_1 = actual_weather.get('actual_weather_1', [])
actual_weather_2 = actual_weather.get('actual_weather_2', [])

# ts 기준으로 오름차순 정렬
sorted_weather_1 = sorted(actual_weather_1, key=lambda x: x['ts'])
sorted_weather_2 = sorted(actual_weather_2, key=lambda x: x['ts'])

# 각 리스트를 24개 구간으로 나누기 위한 간격 계산
step_1 = len(sorted_weather_1) // 24
step_2 = len(sorted_weather_2) // 24

# 결과 출력
print("count | rel_hum  | humidity")
print("-" * 30)

for count in range(1, 25):  # 1부터 24까지
    index_1 = count * step_1 - 1 if step_1 > 0 else 0
    index_2 = count * step_2 - 1 if step_2 > 0 else 0

    # actual_weather_1에서 rel_hum 값 추출
    rel_hum_value = sorted_weather_1[index_1].get('rel_hum') if index_1 < len(sorted_weather_1) else None

    # actual_weather_2에서 humidity 값 추출
    humidity_value = sorted_weather_2[index_2].get('humidity') if index_2 < len(sorted_weather_2) else None

    # 출력
    print(f"{count:5d} | {rel_hum_value:8} | {humidity_value:8}")
```
